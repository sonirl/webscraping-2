```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.flipkart.com/')
```

```python
#clicking_cross_button
crs_btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
crs_btn.click()
```

```python
# #searching sunglasses
srch_bar = driver.find_element_by_xpath("//input[@class='_3704LK']")
srch_bar.send_keys('sunglasses')
```

```python
#clicking_search button
srch_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
srch_button.click()
```

```python
brand = []
description = []
price = []
discount = []


for page in range(0,3):

    #scrapping the required details
    #brand name
    brand_cls = driver.find_elements_by_class_name("_2WkVRV")
    for i in brand_cls[0:40]:
        brand.append(i.text)
    
    #description
    desc_tag = driver.find_elements_by_class_name("IRpwTa")
    for i in desc_tag[0:40]:
        description.append(i.text)
    
    #price
    prc_tag = driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in prc_tag[0:40]:
        price.append(i.text)
        
    #discount
    dis_tag = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span")
    for i in dis_tag[0:40]:
        discount.append(i.text)
        
    next_page=driver.find_elements_by_xpath("//a[@class='ge-49M']")#scraping the list of buttons from the page
    driver.get(next_page[page].get_attribute('href'))#getting the link from the list for next page
```

```python
sol6_data = pd.DataFrame({
    'Brand':brand[0:100],
    'description': description[0:100],
    'price':price[0:100],
    'discount': discount[0:100]
})
```

### Q8 Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.flipkart.com/')
```

```python
#clicking_cross_button
crs_btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
crs_btn.click()
```

```python
#searching sunglasses
srch_bar = driver.find_element_by_xpath("//input[@class='_3704LK']")
srch_bar.send_keys('sneakers')

#clicking_search button
srch_button = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
srch_button.click()
```

```python
shoe_brand=[]
shoe_description=[]
shoe_price=[]
shoe_discount=[]

#scrapping the required details
for page in range(0,4):#for loop for scrapping 3 page
    brands=driver.find_elements_by_class_name('_2WkVRV')#scraping brands name by class name='_2B_pmu'
    for i in brands:
        shoe_brand.append(i.text)#appending the text in Brand list
    desc=driver.find_elements_by_class_name('IRpwTa')#scraping description by class name = '_2mylT6'
    for i in desc:
        shoe_description.append(i.get_attribute('title'))#appending the description in list
    prices=driver.find_elements_by_xpath("//div[@class='_30jeq3']")# scraping the price from the xpath
    for i in prices[:40]:
        shoe_price.append(i.text)
    disc=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")# scraping the discount from the xpath
    for i in disc:
        shoe_discount.append(i.text)
        
    next_page=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    for i in next_page:
        if i.text == 'NEXT':
            driver.get(i.get_attribute('href'))
```

```python
shoe_data = pd.DataFrame({})
shoe_data['brand'] = shoe_brand[0:100]
shoe_data['desc'] = shoe_description[0:100]
shoe_data['price'] = shoe_price[0:100]
shoe_data['descount'] = shoe_discount[0:100]

shoe_data
```

```python
sol10_data = pd.DataFrame({
    'product':name[0:10],
    'price':price[0:10],
    'rating':rating[0:10]
})

sol10_data
```
